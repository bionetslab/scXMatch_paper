```python
import pandas as pd
import glob
import os
```

```python
dfs = dict()
for f in glob.glob("../evaluation_results/1_4_edist/scxmatch/*.csv"):
    dataset = os.path.basename(f).split("xm_benchmark_results_processed_")[1].split(".")[0]
    dfs[dataset] = pd.read_csv(f).set_index("testgroup")
```

```python
scrnaseq = pd.concat(dfs)
scrnaseq["k"] = 100
scrnaseq["time [min]"] = scrnaseq["time_test"] / 60
```
